6. What are the differences between NumPy arrays and Python lists?
~ NumPy arrays and Python lists serve different purposes in Python. NumPy arrays store elements of the same data type, making them faster and more memory-efficient for numerical operations, especially with large datasets. They support element-wise math, matrix operations, and advanced functions like mean or dot product, ideal for scientific computing. However, their size is fixed, requiring new arrays for resizing. In contrast, Python lists can hold mixed data types, offering flexibility but slower performance and higher memory usage. Lists are easily resizable with append or remove operations but lack built-in support for advanced math, requiring loops for similar tasks. Use NumPy arrays for numerical and scientific tasks, and Python lists for general-purpose, mixed-type data handling.

17. Explain the difference between apply() and map() in Pandas.
~ In Pandas, apply() and map() are used to modify data, but they work differently. apply() is used on both DataFrames and Series, allowing you to apply a function to an entire column or row. This makes it more flexible, especially when working with multiple columns. map(), on the other hand, works only on Series and applies a function element-wise to each value. While map() is simpler, apply() is more versatile for complex operations. In short, use map() for basic transformations on individual columns and apply() for more advanced manipulations in a DataFrame.

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data
data = {'A': [1, 2, 3, 4, 5],
        'B': [5, 4, 3, 2, 1],
        'C': [2, 3, 4, 5, 6]}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Display the plot
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#7. Write a program to perform element-wise multiplication of two NumPy arrays
import numpy as np

# Create two NumPy arrays
array1 = np.array([1, 2, 3, 4])
array2 = np.array([5, 6, 7, 8])

# Element-wise multiplication using `*` operator
result1 = array1 * array2

# Element-wise multiplication using `np.multiply()`
result2 = np.multiply(array1, array2)

print("Array 1:", array1)
print("Array 2:", array2)
print("\nMultiplication using `*` operator:", result1)
print("Multiplication using `np.multiply()`:", result2)
```

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Score': [85, 92, 78, 95]}

df = pd.DataFrame(data)

# Define the threshold
threshold = 80

# Filter rows where "Score" is greater than the threshold
filtered_df = df[df['Score'] > threshold]

print("Original DataFrame:")
print(df)

print("\nFiltered DataFrame (Score > 80):")
print(filtered_df)
```

```python
#11. Perform matrix multiplication using NumPy
import numpy as np

# Create two matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Matrix multiplication using np.dot()
result1 = np.dot(matrix1, matrix2)

# Matrix multiplication using @ operator
result2 = matrix1 @ matrix2

print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

print("\nMultiplication using np.dot():")
print(result1)

print("\nMultiplication using @ operator:")
print(result2)
```

```python
#13. Create a 3D scatter plot using Plotly.
import plotly.graph_objects as go
import numpy as np

# Generate Sample Data
np.random.seed(42)
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
z = np.random.rand(50) * 10

# Create a 3D Scatter Plot
fig = go.Figure()

fig.add_trace(go.Scatter3d(x=x, y=y, z=z, mode='markers',
                           marker=dict(size=5, color=z, colorscale='Viridis', opacity=0.8)))

# Add Labels & Title
fig.update_layout(title="3D Scatter Plot Example",
                  scene=dict(xaxis_title="X Axis", yaxis_title="Y Axis", zaxis_title="Z Axis"))

# Show the plot
fig.show()
```
